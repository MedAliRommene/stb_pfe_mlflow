```python
import numpy as np
# Data for each year
years = [2021, 2022, 2023]
non_performing_loan_ratios = [13.5, 14.9, 12.2]  # Sample values based on trend data
risk_costs_million_dinars = [146, 216.4, 98]  # Cost of risk in million dinars for each year

# Set up bar width and positions for each year
bar_width = 0.35
index = np.arange(len(years))

# Create figure and axes with blue-themed colors
fig, ax1 = plt.subplots(figsize=(10, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# Bar chart for Non-performing loan ratios
bar1 = ax1.bar(index, non_performing_loan_ratios, bar_width, color="royalblue", label="Ratio de crances douteuses (%)")
ax1.set_xlabel("Anne", fontsize=12)
ax1.set_ylabel("Ratio de crances douteuses (%)", color="royalblue", fontsize=12)
ax1.set_title("STB - Ratios de Crances Douteuses et Cot du Risque (2021-2023)", fontsize=14, color="navy")
ax1.set_xticks(index)
ax1.set_xticklabels(years)

# Secondary y-axis for the cost of risk with a second bar chart
ax2 = ax1.twinx()
bar2 = ax2.bar(index + bar_width, risk_costs_million_dinars, bar_width, color="r", label="Cot du risque (Millions de Dinars)")
ax2.set_ylabel("Cot du risque (Millions de Dinars)", color="r", fontsize=12)

# Add legend
ax1.legend([bar1, bar2], ["Ratio de crances douteuses (%)", "Cot du risque (Millions de Dinars)"], loc="upper left", fontsize=10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for each year (values from 2013 to 2023)
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
non_performing_loan_ratios = [15.0, 14.5, 14.0, 13.8, 13.2, 13.5, 12.8, 14.0, 13.5, 14.9, 12.2]  # Sample values based on trend data
risk_costs_million_dinars = [200, 180, 160, 150, 140, 135, 120, 110, 146, 216.4, 98]  # Cost of risk in million dinars for each year

# Set up bar width and positions for each year
bar_width = 0.35
index = np.arange(len(years))

# Create figure and axes with professional style
fig, ax1 = plt.subplots(figsize=(14, 8))
plt.style.use('seaborn-v0_8-whitegrid')

# Bar chart for Non-performing loan ratios
bar1 = ax1.bar(index, non_performing_loan_ratios, bar_width, color="#003366", label="Ratio de crances douteuses (%)")

# Annotate bars with the ratio values
for i, v in enumerate(non_performing_loan_ratios):
    ax1.text(i, v + 0.2, str(v), ha='center', fontsize=12, fontweight='bold', color='black')

# Setting the labels and title with more spacing and clear fonts
ax1.set_xlabel("Anne", fontsize=16, fontweight='bold')
ax1.set_ylabel("Ratio de crances douteuses (%)", color="#003366", fontsize=14, fontweight='bold')
ax1.set_title("STB - Ratios de Crances Douteuses et Cot du Risque (2013-2023)", fontsize=18, color="#003366", fontweight='bold')

# Setting the ticks and labels for the x-axis
ax1.set_xticks(index)
ax1.set_xticklabels(years, fontsize=12, fontweight='bold', rotation=45, ha="right")

# Secondary y-axis for the cost of risk with a second bar chart
ax2 = ax1.twinx()
bar2 = ax2.bar(index + bar_width, risk_costs_million_dinars, bar_width, color="#6699CC", label="Cot du risque (Millions de Dinars)")

# Annotate bars with the cost of risk values
for i, v in enumerate(risk_costs_million_dinars):
    ax2.text(i + bar_width, v + 5, str(v), ha='center', fontsize=12, fontweight='bold', color='black')

# Setting the secondary y-axis label
ax2.set_ylabel("Cot du risque (Millions de Dinars)", color="#6699CC", fontsize=14, fontweight='bold')

# Add legend with better positioning
ax1.legend([bar1, bar2], ["Ratio de crances douteuses (%)", "Cot du risque (Millions de Dinars)"], loc="upper left", fontsize=12, frameon=False)

# Remove gridlines for a cleaner look and ensure that only necessary lines are visible
ax1.grid(False)
ax2.grid(False)

# Customize layout to make it tight and avoid clipping
plt.tight_layout()

# Display the plot
plt.show()
```
