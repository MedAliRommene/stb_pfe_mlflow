```python
import numpy as np
# Data for each year
years = [2021, 2022, 2023]
non_performing_loan_ratios = [13.5, 14.9, 12.2]  # Sample values based on trend data
risk_costs_million_dinars = [146, 216.4, 98]  # Cost of risk in million dinars for each year

# Set up bar width and positions for each year
bar_width = 0.35
index = np.arange(len(years))

# Create figure and axes with blue-themed colors
fig, ax1 = plt.subplots(figsize=(10, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# Bar chart for Non-performing loan ratios
bar1 = ax1.bar(index, non_performing_loan_ratios, bar_width, color="royalblue", label="Ratio de créances douteuses (%)")
ax1.set_xlabel("Année", fontsize=12)
ax1.set_ylabel("Ratio de créances douteuses (%)", color="royalblue", fontsize=12)
ax1.set_title("STB - Ratios de Créances Douteuses et Coût du Risque (2021-2023)", fontsize=14, color="navy")
ax1.set_xticks(index)
ax1.set_xticklabels(years)

# Secondary y-axis for the cost of risk with a second bar chart
ax2 = ax1.twinx()
bar2 = ax2.bar(index + bar_width, risk_costs_million_dinars, bar_width, color="r", label="Coût du risque (Millions de Dinars)")
ax2.set_ylabel("Coût du risque (Millions de Dinars)", color="r", fontsize=12)

# Add legend
ax1.legend([bar1, bar2], ["Ratio de créances douteuses (%)", "Coût du risque (Millions de Dinars)"], loc="upper left", fontsize=10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for each year
years = [2021, 2022, 2023]
non_performing_loan_ratios = [13.5, 14.9, 12.2]  # Sample values based on trend data
risk_costs_million_dinars = [146, 216.4, 98]  # Cost of risk in million dinars for each year

# Set up bar width and positions for each year
bar_width = 0.35
index = np.arange(len(years))

# Create figure and axes with blue-themed colors
fig, ax1 = plt.subplots(figsize=(10, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# Bar chart for Non-performing loan ratios
bar1 = ax1.bar(index, non_performing_loan_ratios, bar_width, color="#003366", label="Ratio de créances douteuses (%)")
ax1.set_xlabel("Année", fontsize=14, fontweight='bold')
ax1.set_ylabel("Ratio de créances douteuses (%)", color="#003366", fontsize=14, fontweight='bold')
ax1.set_title("STB - Ratios de Créances Douteuses et Coût du Risque (2021-2023)", fontsize=16, color="#003366", fontweight='bold')
ax1.set_xticks(index)
ax1.set_xticklabels(years, fontsize=12, fontweight='bold')

# Secondary y-axis for the cost of risk with a second bar chart
ax2 = ax1.twinx()
bar2 = ax2.bar(index + bar_width, risk_costs_million_dinars, bar_width, color="#6699CC", label="Coût du risque (Millions de Dinars)")
ax2.set_ylabel("Coût du risque (Millions de Dinars)", color="#6699CC", fontsize=14, fontweight='bold')

# Add legend
ax1.legend([bar1, bar2], ["Ratio de créances douteuses (%)", "Coût du risque (Millions de Dinars)"], loc="upper left", fontsize=12, frameon=False)

# Remove gridlines for a cleaner look
ax1.grid(False)
ax2.grid(False)

# Display the plot
plt.tight_layout()
plt.show()
```
